## Arquitectura de una red

La topología de una red queda especificada por 3 características principales:

1. El número de **capas** a utilizar.

2. Si la red permite o no el fenómeno de **travel backward**.

3. El número de nodos en cada una de las capas de la red.

### a. Número de capas 

Las neuronas se organizan en un ordenamiento sucesivo, en capas.

### b. Caso básico de 1 capa (modelo perceptrón)

Desde las señales de entrada, una primera capa recibe cada las señales de entrada. En esta capa, una neurona por señal es la encargada de su procesamiento, denominandose estas **nodos de entrada**, y la capa **capa de entrada**. El procesamiento de la señal de entrada se realiza en base a la función de activación de la red. Cada uno de estos nodos envía una señal de salida a un **nodo de salida**, que mediante su función de activación genera finalmente una predicción.

Esta topología describe una **red de una capa**, que hace referencia a la capa de entrada. Este tipo de redes básicas es suficiente para conjuntos de datos linealmente separables.

La forma básica de complejizar esta red es *agregando capas* entre la capa de entrada y la capa de salida. Las capas que se encuentran entre estas dos capas se suelen denominar **capas ocultas (hidden layers)**. En el caso de redes de una capa, el número de capas ocultas es cero.

Una red neuronal con múltiples capas ocultas se suele denominar **red neuronal profunda (Deep Neural Network - DNN)**, y el campo de estudio de este tipo de modelos se denomina **Deep Learning**.

### c. Flujo de la información (posibilidad de travel backward)

La topología también está definida por la posibilidad de restringir la direccionalidad del flujo de la información.

El entrenamiento de la red puede realizarse sólo permitiendo que la información de entrada *viaje* a los nodos de salida respetando siempre esa dirección, o permitiendo (mediante loops), que se dezplace en las dos direcciones posibles (hacia adelante y hacia atrás). En el primer caso se trata de redes de tipo **feedforward**, mientras que en el último se denominan de tipo **recurrent o feedback**. 

Además de esto, el flujo de información puede tener la propiedad de **delay**, o sea la propiedad de *mantener registro de las operaciones realizadas en un conjunto de pasos* ^[chequear esta definición.].

En la práctica, se suelen utilizar sólo redes de tipo *feedforward*.

El estándar de trabajo de redes neuronales es el modelo denominado **Multilayer Perceptron (modelo neuronal MLP)**, que es un modelo multicapa, feedforward.

### d. Cantidad de nodos por capa

La complejidad de la topología de una red también varía en función de la cantidad de nodos por capa.

Por un lado, en la capa de entrada la cantidad de nodos se corresponde necesariamente a la cantidad de variables input del modelo. En la capa de salida la cantida de nodos también se define necesariamente en base a la cantidad de atributos a predecir (para el caso cuantitativo), o por la cantidad de valores de clase (para el caso cualitativo). Los nodos de las capas ocultas, sin embargo, tienen una cantidad de nodos a definir por el diseñador de la red.

La definición de la cantidad de nodos a utilizar en las capas ocultas es un tema que no es sencillo. Como en todos este tipo de problemas, el aumento de la cantida de nodos implica una mejora de perfomance sobre datos de entrenamiento, con el correspondiente riesgo de sobreajuste. 

Además, agregar nodos a una red es computacionalmente costoso.

Se ha demostrado (Lantz (2015)), que una red con una capa oculta y un número suficiente de neuronas permite cumplir el rol de *aproximar cualquier tipo de función en un intervalo finito con una precisión arbitraria*.
