## E. Un ejemplo de red neuronal básica: el perceptrón

Se denomina **modelo perceptron** a una red neuronal que sólo posee capa de entrada y capa de salida, esta última de una sola neurona.

El modelo perceptrón realiza una predicción en base a la suma ponderada de los valores de las variables de entrada, más un **término de sesgo $w_0$**. O sea tenemos que el modelo viene dado por:

$$\hat{y} = w_0 + w_1*x_1 + ...+w_p*x_p = X*W$$

Como se ve, es un modelo lineal.

El proceso de aprendiza consiste en **ajustar los parámetros $w_i$**, de manera que los valores de respuesta del modelo $\hat{y}$ sean consistentes con los verdaderos $y$.

Este proceso de ajuste se realiza mediante iteraciones sucesivas. Al inicio, el algoritmo establece un valor aleatorio a los parámetros. A partir de los resultados, el proceso de ajuste de los parámetros viene dado por la siguiente **fórmula de actualización**:

$$w_j^{(k+1)} = w_j^{(k)} + \lambda(y_i - \hat{y}_i)*x_{ij}$$

donde el término $\lambda$ se denomina **tasa de aprendizaje**, toma valores entre cero y uno y se utiliza para controlar la fuerza del ajuste a realizar ( de 0 a 1, menor a mayor ajuste)

### Algoritmo del modelo perceptrón

Se establecen acá los pasos que realiza el modelo perceptrón. Observar que el modelo perceptrón aquí planteado está pensado como realizando tareas de regresión, pero que en la foto está planteado como modelo de clasificación (simplemente depende de la función de activación!)

El algoritmo perceptrón es el siguiente: 

+ Supongo dados una matriz de datos $X$ y su respuesta $y$.
+ Supongo definidos hiperparámetros $\alpha$ = tasa de aprendizaje y MAX_ITER = iteraciones máximas del algoritmo para encontrar una respuesta.
+ Supongo definida una función de activación $f$.
+ Paso 1: Defino valores aleatorios del vector de pesos $W$, vector de $k + 1$ componentes donde $k$ es la cantidad de atributos de $X$.
+ Paso 2: (Inicio de loop para cada registro n): Calculo la **Señal neta**. $NETA = W \cdot X_n$.
+ Paso 3: (Sigue loop). Calculo respuesta en base a **función de activación** $y_{pred} = f(NETA)$.
+ Paso 4: (Sigue loop). Actualizo vector de pesos mediante función de aprendizaje $W_{nuevo} = W + \alpha(y_{pred} - y_e)\cdot X_e$. 
+ Una iteración es la repetición de estos pasos para cada registro, en orden. Termina el proceso al alcanzar la cantidad de iteraciones MAX_ITER, o cuando no exista variación en el valor de W.
